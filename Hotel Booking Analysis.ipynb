### Overview of reserved room type and assigned room type

````python
# Selecting three colums from DataFrame which are: "is_canceled", "reserved_room_type", "assigned_room_type"
df[['is_canceled','reserved_room_type','assigned_room_type']]
````

````python
# Define a function to check whether a customer get their preferred reserved room type
def rrt_art_same(df):
  '''
  This function returns whether the customer get their preferred reserved room type or not.
  '''
  if df['reserved_room_type'] == df['assigned_room_type']:
      return 'True'
  else:
      return 'False'
````

````python
# Apply to dataframe
df['rrt_art_same'] = df.apply(rrt_art_same,axis=1)

# Check dataframe
df[['is_canceled','reserved_room_type','assigned_room_type','rrt_art_same']]
````

````python
# Counting total number of different types of assigned and reserved room type and creating a DataFrame
p1 = df['reserved_room_type'].value_counts().reset_index()
p2 = df['assigned_room_type'].value_counts().reset_index()
````

````python
# Merging both DataFrame using right join on room type
merged_RJ_df = pd.merge(p1, p2, how = 'right', left_on='index', right_on='index')

# Filling the Missing value with 0
merged_RJ_df.fillna(0, inplace = True)

# Setting the index to room type
merged_RJ_df = merged_RJ_df.set_index('index')
````

````python
# Plotting the number of customers who reserved their different room types and comparing it with nuber of assigned customer with different room type
plt.figure(figsize = (20,10), dpi=110)
plt.plot(merged_RJ_df['reserved_room_type'],'go--', linewidth=2, markersize=4)
plt.plot(merged_RJ_df['assigned_room_type'],'ro-', linewidth=2, markersize=4)
plt.legend(title='Room Type', labels=['Reserved room','Assigned room'])
plt.title('Comparison between Reserved Room and Assigned Room', weight = 'bold')
plt.xlabel('Room Type')
plt.ylabel('Count')
````

Room Type A is the most preferred and assigned room and Room type L and P is least preferred and assigned room.
